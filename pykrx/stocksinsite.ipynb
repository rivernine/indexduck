```python
# pykrx to mysql

from sqlalchemy import create_engine
import pymysql
import pandas as pd
from pykrx import stock
import time
import os
from openpyxl import load_workbook

try:
  ticker_name = {}
  for ticker in stock.get_market_ticker_list():
    name = stock.get_market_ticker_name(ticker)
    ticker_name[ticker] = name
  for ticker in stock.get_market_ticker_list(market='KOSDAQ'):
    name = stock.get_market_ticker_name(ticker)
    ticker_name[ticker] = name

  start = "20210101"
  end = "20210902"

  for ticker in ticker_name:
    try:
      df = stock.get_market_ohlcv_by_date(start, end, ticker).drop(['시가', '고가', '저가'], axis=1).rename(columns={'종가':'close', '거래량':'volume'})
      df.index.names = ['date']
      df_volume_netbuy = stock.get_market_trading_volume_by_date(start, end, ticker, on="순매수").rename(columns={'기관합계':'insVol', '기타법인':'etcVol', '개인':'indVol', '외국인합계':'forVol'}).drop(['전체'], axis=1)
      df_amount_netbuy = stock.get_market_trading_value_by_date(start, end, ticker, on="순매수").rename(columns={'기관합계':'insAmt', '기타법인':'etcAmt', '개인':'indAmt', '외국인합계':'forAmt'}).drop(['전체'], axis=1)
      df_total = pd.concat([df, df_volume_netbuy, df_amount_netbuy], axis = 1, join='inner')
      df_total.insert(0, 'ticker', str(ticker))
      df_total.insert(1, 'name', ticker_name[ticker])

      db_connection = create_engine('mysql+pymysql://root:test@192.168.56.100:3306/INDEX_DUCK', encoding='utf-8')
      conn = db_connection.connect()

      df_total.to_sql(name='stocks', con=db_connection, if_exists='append', index=True, index_label="date")

      conn.close()
      time.sleep(1)
    except Exception as e:
      print(e)
      pass
except Exception as ex:
  print(ex)
```

```python
# 상관계수 구하기
import pandas as pd
import pymysql
from sqlalchemy import create_engine
from pykrx import stock

write_connection = create_engine('mysql+pymysql://root:root@192.168.56.100:3306/INDEX_DUCK', encoding='utf-8')
conn = write_connection.connect()

db = pymysql.connect(host='192.168.56.100', port=3306, user='root', passwd='root', db='INDEX_DUCK', charset='utf8', cursorclass=pymysql.cursors.DictCursor)
cursor = db.cursor()

ticker_cap = {}

capDf = stock.get_market_cap_by_ticker("20200625")
for ticker in capDf.index:
  ticker_cap[str(ticker)] = capDf.loc[ticker, '시가총액']

period = 'last_3m'
# period = 'last_6m'
# period = 'last_12m'
try:
  getTickersSql = "SELECT DISTINCT ticker FROM %s" % (period)
  cursor.execute(getTickersSql)
  tickers = cursor.fetchall()
  for ticker in tickers:
    print(ticker)
    getTickerInfoSql = "SELECT * FROM %s WHERE ticker = '%s'" % (period, ticker['ticker'])
    cursor.execute(getTickerInfoSql)
    df = pd.DataFrame(cursor.fetchall())
    name = df.loc()[:, 'name'].iloc[0]
    if len(df) < 1:
      continue

    corrDf = df.corr().loc()[:, ['insVolCum', 'etcVolCum', 'indVolCum', 'forVolCum', 'insAmtCum', 'etcAmtCum', 'indAmtCum', 'forAmtCum']].iloc[:1]
    corrDf.insert(0, 'ticker', ticker['ticker'])
    corrDf.insert(1, 'name', name)
    corrDf.insert(2, 'period', period)
    corrDf.insert(3, 'cap', ticker_cap[ticker['ticker']])
    try:
      corrDf.to_sql(name="correl", con=write_connection, if_exists='append', index=False)
      db.commit()
    except Exception as ex1:
      print('ex1', ex1)
      pass
except Exception as ex2:
  print('ex2', ex2)
  pass

db.close()
conn.close()
```
