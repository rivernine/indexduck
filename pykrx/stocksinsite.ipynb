```python
# 누적 순매수 계산
import pandas as pd
import pymysql
from datetime import datetime, timedelta

ADDR = '192.168.56.100'
PORT = '3306'
DB = 'INDEX_DUCK'
ID = 'root'
PW = 'test'

periods = [['last_1m', 1], ['last_3m', 3], ['last_6m', 6], ['last_12m', 12]]

### Drop & Create table
try:
  db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8', cursorclass=pymysql.cursors.DictCursor)
  cursor = db.cursor()
  dropTableSql = "DROP TABLE last_1m, last_3m, last_6m, last_12m"
  cursor.execute(dropTableSql)
except Exception as e:
  print(e)
  pass
for period in periods:
  createTableSql = """
    CREATE TABLE %s  (
      date        date,
      ticker      varchar(10),
      name        varchar(255),
      close       bigint,
      volume      bigint,
      insVolCum   bigint,
      etcVolCum   bigint,
      indVolCum   bigint,
      forVolCum   bigint,
      insAmtCum   bigint,
      etcAmtCum   bigint,
      indAmtCum   bigint,
      forAmtCum   bigint,
      primary key (ticker, date)
    ) engine = InnoDB;
  """ % (period[0])
  cursor.execute(createTableSql)
db.commit()
db.close()

db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8')
cursor = db.cursor()
try:
  getTickersSql = "SELECT DISTINCT ticker FROM stocks"
  cursor.execute(getTickersSql)
  tickers = list(cursor.fetchall())


  for period in periods:
    date = str((datetime.now() - timedelta(weeks=period[1]*4)).date())
    for ticker in tickers:
      print(ticker)
      getTickerInfoSql = "SELECT * FROM stocks WHERE ticker = '" + ticker[0] + "' AND date > '" + date + "'"
      cursor.execute(getTickerInfoSql)
      tickerInfos = list(map(list, cursor.fetchall()))
      if len(tickerInfos) < 1:
        continue

      tickerCums = [tickerInfos[0]]
      for idx in range(1, len(tickerInfos)):
        cur = tickerInfos[idx]
        bef = tickerCums[idx - 1]
        cum = [a + b for a, b in zip(bef[5:], cur[5:])]
        tickerCums.append(cur[:5] + cum)
      try:
        updateInfoSql = "REPLACE INTO " + period[0] + "(date, ticker, name, close, volume, insVolCum, etcVolCum, indVolCum, forVolCum, insAmtCum, etcAmtCum, indAmtCum, forAmtCum) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        cursor.executemany(updateInfoSql, tickerCums)
        db.commit()
      except Exception as ex1:
        print('ex1', ex1)
        pass
except Exception as ex2:
  print('ex2', ex2)
  pass

db.close()
```

```python
# 상관계수 구하기
import pandas as pd
import pymysql
from sqlalchemy import create_engine

ADDR = '192.168.56.100'
PORT = '3306'
DB = 'INDEX_DUCK'
ID = 'root'
PW = 'test'

ticker_cap, ticker_fund = {}, {}
periods = ['last_1m', 'last_3m', 'last_6m', 'last_12m']

### Drop & Create table
db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8', cursorclass=pymysql.cursors.DictCursor)
cursor = db.cursor()
dropTableSql = "DROP TABLE correl"
cursor.execute(dropTableSql)
createTableSql = """
  CREATE TABLE correl(
  ticker      varchar(10),
  name        varchar(255),
  period      varchar(10),
  insVolCum   double,
  etcVolCum   double,
  indVolCum   double,
  forVolCum   double,
  insAmtCum   double,
  etcAmtCum   double,
  indAmtCum   double,
  forAmtCum   double,
  primary key (ticker, period)
) engine = InnoDB;
"""
cursor.execute(createTableSql)
db.commit()
db.close()

### Calculate correl 
for period in periods:
  try:
    db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8', cursorclass=pymysql.cursors.DictCursor)
    cursor = db.cursor()

    getTickersSql = "SELECT DISTINCT ticker FROM %s" % (period)
    cursor.execute(getTickersSql)
    tickers = cursor.fetchall()
    for ticker in tickers:
      t = ticker['ticker']
      print(t)
      getTickerInfoSql = "SELECT * FROM %s WHERE ticker = '%s'" % (period, t)
      cursor.execute(getTickerInfoSql)
      df = pd.DataFrame(cursor.fetchall())
      name = df.loc()[:, 'name'].iloc[0]
      if len(df) < 1:
        continue

      corrDf = df.corr().loc()[:, ['insVolCum', 'etcVolCum', 'indVolCum', 'forVolCum', 'insAmtCum', 'etcAmtCum', 'indAmtCum', 'forAmtCum']].iloc[:1]
      corrDf.insert(0, 'ticker', t)
      corrDf.insert(1, 'name', name)
      corrDf.insert(2, 'period', period)
      try:
        write_connection = create_engine('mysql+pymysql://'+ID+':'+PW+'@'+ADDR+':'+PORT+'/'+DB, encoding='utf-8', pool_pre_ping=True).connect()
        corrDf.to_sql(name="correl", con=write_connection, if_exists='append', index=False)
      except Exception as ex1:
        print('ex1', ex1)
        pass
      finally:
        write_connection.close()
  except Exception as ex2:
    print('ex2', ex2)
    db.close()  
    pass
```

```python
# 티커 별 정보 셋팅
from sqlalchemy import create_engine
import pymysql
import pandas as pd
from pykrx import stock
import time
from openpyxl import load_workbook
from datetime import datetime, timedelta

ADDR = '192.168.56.100'
PORT = '3306'
DB = 'INDEX_DUCK'
ID = 'root'
PW = 'root'

### Drop & Create table
try:
  db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8', cursorclass=pymysql.cursors.DictCursor)
  cursor = db.cursor()
  dropTableSql = "DROP TABLE stocks_info"
  cursor.execute(dropTableSql)
except Exception as ex:
  print(ex)
  pass
createTableSql = """
  CREATE TABLE stocks_info  (
    ticker      varchar(10),
    name        varchar(255),
    market      varchar(10),
    cap         bigint,
    bps         double,
    per         double,
    pbr         double,
    eps         double,
    divi        double,
    dps         double,
    primary key (ticker, name)
  ) engine = InnoDB;
"""
cursor.execute(createTableSql)
db.commit()
db.close()

try:
  ticker_name, ticker_cap, ticker_fund = {}, {}, {}

  # 공휴일을 제외한 최근날짜 입력
  capDf = stock.get_market_cap_by_ticker("20211001").loc[:, '시가총액']
  fundDf = stock.get_market_fundamental_by_ticker("20211001", market="ALL")
  # capDf = stock.get_market_cap_by_ticker(str(datetime.now().date()).replace('-', '')).loc[:, '시가총액']
  # fundDf = stock.get_market_fundamental_by_ticker(str(datetime.now().date()).replace('-', ''), market="ALL")
  for ticker in capDf.index:
    ticker_cap[ticker] = capDf.loc[ticker]
  for ticker in fundDf.index:
    ticker_fund[ticker] = list(fundDf.loc[ticker])

  for ticker in stock.get_market_ticker_list():
    name = stock.get_market_ticker_name(ticker)
    ticker_name[ticker] = [name, 'KOSPI']
    
  for ticker in stock.get_market_ticker_list(market='KOSDAQ'):
    name = stock.get_market_ticker_name(ticker)
    ticker_name[ticker] = [name, 'KOSDAQ']

  for ticker in ticker_name:
    if ticker not in capDf.index or ticker not in fundDf.index:
      continue
    print(ticker)
    tickerInfo = [ticker]
    tickerInfo += ticker_name[ticker]
    tickerInfo.append(ticker_cap[ticker])
    tickerInfo += ticker_fund[ticker]

    tickerInfoDf = pd.DataFrame([tickerInfo], columns = ['ticker', 'name', 'market', 'cap', 'bps', 'per', 'pbr', 'eps', 'divi', 'dps'])
    try:
      write_connection = create_engine('mysql+pymysql://'+ID+':'+PW+'@'+ADDR+':'+PORT+'/'+DB, encoding='utf-8', pool_pre_ping=True).connect()
      tickerInfoDf.to_sql(name="stocks_info", con=write_connection, if_exists='append', index=False)
    except Exception as ex1:
      print('ex1', ex1)
      pass
    finally:
      write_connection.close()
except Exception as ex:
  print(ex)
```

```python
# 정규화 작업
import pandas as pd
import pymysql
from datetime import datetime, timedelta
from sqlalchemy import create_engine
from sklearn.preprocessing import MinMaxScaler

ADDR = '192.168.56.100'
PORT = '3306'
DB = 'INDEX_DUCK'
ID = 'root'
PW = 'test'

periods = ['last_1m', 'last_3m', 'last_6m', 'last_12m']

### Drop & Create table
try:
  db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8', cursorclass=pymysql.cursors.DictCursor)
  cursor = db.cursor()
  dropTableSql = "DROP TABLE last_1m_norm, last_3m_norm, last_6m_norm, last_12m_norm"
  cursor.execute(dropTableSql)
except Exception as e:
  print(e)
  pass
for period in periods:
  createTableSql = """
    CREATE TABLE %s  (
      date            date,
      ticker          varchar(10),
      name            varchar(255),
      closeNorm       double,
      volume          bigint,
      insVolCumNorm   double,
      etcVolCumNorm   double,
      indVolCumNorm   double,
      forVolCumNorm   double,
      insAmtCumNorm   double,
      etcAmtCumNorm   double,
      indAmtCumNorm   double,
      forAmtCumNorm   double,
      primary key (ticker, date)
    ) engine = InnoDB;
  """ % (period + '_norm')
  cursor.execute(createTableSql)
db.commit()
db.close()

scaler = MinMaxScaler(feature_range=(0, 1))
try:
  for period in periods:
      db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8')
      cursor = db.cursor()
      getTickersSql = "SELECT DISTINCT ticker FROM " + period
      cursor.execute(getTickersSql)
      tickers = list(cursor.fetchall())
      db.close()
      for ticker in tickers:
        # date|ticker|name|close|volume|insVolCum|etcVolCum|indVolCum|forVolCum|insAmtCum|etcAmtCum|indAmtCum|forAmtCum
        try:
          db = pymysql.connect(host=ADDR, port=int(PORT), user=ID, passwd=PW, db=DB, charset='utf8')
          cursor = db.cursor()
          getDataSql = "SELECT * FROM " + period + " WHERE ticker = '" + ticker[0] + "'"
          cursor.execute(getDataSql)
          df = pd.DataFrame(cursor.fetchall())
          tdf = df.copy()
          df[[3, 5, 6, 7, 8, 9, 10, 11, 12]] = scaler.fit_transform(tdf[[3, 5, 6, 7, 8, 9, 10, 11, 12]])
          
          df.columns = ['date', 'ticker', 'name', 'closeNorm', 'volume', 'insVolCumNorm', 'etcVolCumNorm', 'indVolCumNorm', 'forVolCumNorm', 'insAmtCumNorm', 'etcAmtCumNorm', 'indAmtCumNorm', 'forAmtCumNorm']
          write_connection = create_engine('mysql+pymysql://'+ID+':'+PW+'@'+ADDR+':'+PORT+'/'+DB, encoding='utf-8', pool_pre_ping=True).connect()
          df.to_sql(name=period + "_norm", con=write_connection, if_exists='append', index=False)
        except Exception as ex1:
          print('ex1', ex1)
          pass
        finally:
          write_connection.close()
          db.close()
except Exception as ex2:
  print('ex2', ex2)
  pass 
```
