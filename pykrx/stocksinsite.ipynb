```python
from pykrx import stock
from dataclasses import dataclass
import time

@dataclass
class Info:
  close: int = None
  volume_buy: int = None
  volume_sell: int = None
  volume_netbuy: int = None
  amount_buy: int = None
  amount_sell: int = None
  amount_netbuy: int = None
  individual_volume_buy: int = None
  individual_volume_sell: int = None
  individual_volume_netbuy: int = None
  individual_amount_buy: int = None
  individual_amount_sell: int = None
  individual_amount_netbuy: int = None
  foreign_volume_buy: int = None
  foreign_volume_sell: int = None
  foreign_volume_netbuy: int = None
  foreign_amount_buy: int = None
  foreign_amount_sell: int = None
  foreign_amount_netbuy: int = None
  institution_volume_buy: int = None
  institution_volume_sell: int = None
  institution_volume_netbuy: int = None
  institution_amount_buy: int = None
  institution_amount_sell: int = None
  institution_amount_netbuy: int = None
  corporation_volume_buy: int = None
  corporation_volume_sell: int = None
  corporation_volume_netbuy: int = None
  corporation_amount_buy: int = None
  corporation_amount_sell: int = None
  corporation_amount_netbuy: int = None

ticker_name, ticker_info = {}, {}
for ticker in stock.get_market_ticker_list()[:5]:
  name = stock.get_market_ticker_name(ticker)
  ticker_name[ticker] = name
  ticker_info[ticker] = {}

start = "20210825"
end = "20210830"

for ticker in ticker_name:
  df_close = stock.get_market_ohlcv_by_date(start, end, ticker)
  # 날짜(index) : 기관합계 / 기타법인 / 개인 / 외국인합계 / 전체
  df_volume_buy = stock.get_market_trading_volume_by_date(start, end, ticker, on="매수")
  df_volume_sell = stock.get_market_trading_volume_by_date(start, end, ticker, on="매도")
  df_volume_netbuy = stock.get_market_trading_volume_by_date(start, end, ticker, on="순매수")
  time.sleep(1)
  df_amount_buy = stock.get_market_trading_value_by_date(start, end, ticker, on="매수")
  df_amount_sell = stock.get_market_trading_value_by_date(start, end, ticker, on="매도")
  df_amount_netbuy = stock.get_market_trading_value_by_date(start, end, ticker, on="순매수")
  
  for idx in df_close.index:
    date = str(idx).split(' ')[0].replace('-', '')
    ticker_info[ticker][date] = Info()
    ticker_info[ticker][date].close = df_close.loc[idx, "종가"]
    ticker_info[ticker][date].institution_volume_buy, ticker_info[ticker][date].corporation_volume_buy, ticker_info[ticker][date].individual_volume_buy, ticker_info[ticker][date].foreign_volume_buy, ticker_info[ticker][date].volume_buy = df_volume_buy.loc[idx] 
    ticker_info[ticker][date].institution_volume_sell, ticker_info[ticker][date].corporation_volume_sell, ticker_info[ticker][date].individual_volume_sell, ticker_info[ticker][date].foreign_volume_sell, ticker_info[ticker][date].volume_sell = df_volume_sell.loc[idx] 
    ticker_info[ticker][date].institution_volume_netbuy, ticker_info[ticker][date].corporation_volume_netbuy, ticker_info[ticker][date].individual_volume_netbuy, ticker_info[ticker][date].foreign_volume_netbuy, ticker_info[ticker][date].volume_netbuy = df_volume_netbuy.loc[idx] 

    ticker_info[ticker][date].institution_amount_buy, ticker_info[ticker][date].corporation_amount_buy, ticker_info[ticker][date].individual_amount_buy, ticker_info[ticker][date].foreign_amount_buy, ticker_info[ticker][date].amount_buy = df_amount_buy.loc[idx] 
    ticker_info[ticker][date].institution_amount_sell, ticker_info[ticker][date].corporation_amount_sell, ticker_info[ticker][date].individual_amount_sell, ticker_info[ticker][date].foreign_amount_sell, ticker_info[ticker][date].amount_sell = df_amount_sell.loc[idx] 
    ticker_info[ticker][date].institution_amount_netbuy, ticker_info[ticker][date].corporation_amount_netbuy, ticker_info[ticker][date].individual_amount_netbuy, ticker_info[ticker][date].foreign_amount_netbuy, ticker_info[ticker][date].amount_netbuy = df_amount_netbuy.loc[idx] 

  # for idx in df.index:
  #   time.sleep(1)
  #   date = str(idx).split(' ')[0].replace('-', '')
  #   ticker_info[ticker][date] = Info()
  #   ticker_info[ticker][date].close = df.loc[idx, "종가"]
  #   ticker_info[ticker][date].institution
  #   df_buy.loc[idx]

  #   df_volume = stock.get_market_trading_volume_by_investor(date, date, ticker)    
  #   individual_volume = df_volume.loc['개인']
  #   foreign_volume = df_volume.loc['외국인']
  #   pension_volume = df_volume.loc['연기금 등']
  #   institutation_volume = df_volume.loc['기관합계'] - pension_volume
  #   volume = df_volume.loc['전체']
  #   ticker_info[ticker][date].pension_volume_sell, ticker_info[ticker][date].pension_volume_buy, ticker_info[ticker][date].pension_volume_netbuy = pension_volume
  #   ticker_info[ticker][date].institutation_volume_sell, ticker_info[ticker][date].institutation_volume_buy, ticker_info[ticker][date].institutation_volume_netbuy = institutation_volume
  #   ticker_info[ticker][date].individual_volume_sell, ticker_info[ticker][date].individual_volume_buy, ticker_info[ticker][date].individual_volume_netbuy = individual_volume
  #   ticker_info[ticker][date].foreign_volume_sell, ticker_info[ticker][date].foreign_volume_buy, ticker_info[ticker][date].foreign_volume_netbuy = foreign_volume
  #   ticker_info[ticker][date].volume_sell, ticker_info[ticker][date].volume_buy, ticker_info[ticker][date].volume_netbuy = volume

  #   df_amount = stock.get_market_trading_value_by_investor(date, date, ticker)   
  #   individual_amount = df_amount.loc['개인']
  #   foreign_amount = df_amount.loc['외국인']
  #   pension_amount = df_amount.loc['연기금 등']
  #   institutation_amount = df_amount.loc['기관합계'] - pension_amount
  #   amount = df_amount.loc['전체']
  #   ticker_info[ticker][date].pension_amount_sell, ticker_info[ticker][date].pension_amount_buy, ticker_info[ticker][date].pension_amount_netbuy = pension_amount
  #   ticker_info[ticker][date].institutation_amount_sell, ticker_info[ticker][date].institutation_amount_buy, ticker_info[ticker][date].institutation_amount_netbuy = institutation_amount
  #   ticker_info[ticker][date].individual_amount_sell, ticker_info[ticker][date].individual_amount_buy, ticker_info[ticker][date].individual_amount_netbuy = individual_amount
  #   ticker_info[ticker][date].foreign_amount_sell, ticker_info[ticker][date].foreign_amount_buy, ticker_info[ticker][date].foreign_amount_netbuy = foreign_amount
  #   ticker_info[ticker][date].amount_sell, ticker_info[ticker][date].amount_buy, ticker_info[ticker][date].amount_netbuy, = amount

ticker_info
```
