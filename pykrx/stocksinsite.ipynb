```python
from pykrx import stock
# 티커리스트 조회
# input : 일자(미 입력 시 최근 영업일 DEFAULT)
#   - market: KOSPI(DEFAULT), KOSDAQ, KONEX, ALL
# output: 코스피 시장에 상장된 ticker list
tickers = stock.get_market_ticker_list("20210830")
```

```python
# 종목이름 조회
# input : ticker
# output: 종목이름
for ticker in stock.get_market_ticker_list():
  name = stock.get_market_ticker_name(ticker)
  name
```

```python
# 일자별 OHLCV 조회
# input : 시작일/종료일/티커/(Optional)(d/m/y)
# output: 일자별 OHLCV DataFrame
df = stock.get_market_ohlcv_by_date("20210829", "20210830", "005930")
df.head(3)
```

```python
# 전체 종목 시세 조회
# input : 날짜
#   - market: KOSPI(DEFAULT), KOSDAQ, KONEX, ALL
# output: 코스피 전 종목 시세 DataFrame
df = stock.get_market_ohlcv_by_ticker("20210830")
df
```
