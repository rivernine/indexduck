```python
from pykrx import stock
from dataclasses import dataclass

# ticker_info[ticker][date]
## 0: close
## 1: volume-buy
## 2: volume-sell
## 3: volume-netbuy
## 4: amount-buy
## 5: amount-sell
## 6: amount-netbuy
## 7: individual-volume-buy
## 8: individual-volume-sell
## 9: individual-volume-netbuy
## 10: individual-amount-buy
## 11: individual-amount-sell
## 12: individual-amount-netbuy
## 13: foreign-volume-buy
## 14: foreign-volume-sell
## 15: foreign-volume-netbuy
## 16: foreign-amount-buy
## 17: foreign-amount-sell
## 18: foreign-amount-netbuy
## 19: institution-volume-buy
## 20: institution-volume-sell
## 21: institution-volume-netbuy
## 22: institution-amount-buy
## 23: institution-amount-sell
## 24: institution-amount-netbuy
## 25: pension-volume-buy
## 26: pension-volume-sell
## 27: pension-volume-netbuy
## 28: pension-amount-buy
## 29: pension-amount-sell
## 30: pension-amount-netbuy
## 31: finance-volume-buy
## 32: finance-volume-sell
## 33: finance-volume-netbuy
## 34: finance-amount-buy
## 35: finance-amount-sell
## 36: finance-amount-netbuy

@dataclass
class Info:
  close: int = None
  volume_buy: int = None
  volume_sell: int = None
  volume_netbuy: int = None
  amount_buy: int = None
  amount_sell: int = None
  amount_netbuy: int = None
  individual_volume_buy: int = None
  individual_volume_sell: int = None
  individual_volume_netbuy: int = None
  individual_amount_buy: int = None
  individual_amount_sell: int = None
  individual_amount_netbuy: int = None
  foreign_volume_buy: int = None
  foreign_volume_sell: int = None
  foreign_volume_netbuy: int = None
  foreign_amount_buy: int = None
  foreign_amount_sell: int = None
  foreign_amount_netbuy: int = None
  institution_volume_buy: int = None
  institution_volume_sell: int = None
  institution_volume_netbuy: int = None
  institution_amount_buy: int = None
  institution_amount_sell: int = None
  institution_amount_netbuy: int = None
  pension_volume_buy: int = None
  pension_volume_sell: int = None
  pension_volume_netbuy: int = None
  pension_amount_buy: int = None
  pension_amount_sell: int = None
  pension_amount_netbuy: int = None
  finance_volume_buy: int = None
  finance_volume_sell: int = None
  finance_volume_netbuy: int = None
  finance_amount_buy: int = None
  finance_amount_sell: int = None
  finance_amount_netbuy: int = None

ticker_name, ticker_info = {}, {}
for ticker in stock.get_market_ticker_list():
  name = stock.get_market_ticker_name(ticker)
  ticker_name[ticker] = name
  ticker_info[ticker] = {}

for ticker in ticker_name:
  df = stock.get_market_ohlcv_by_date("20210830", "20210830", ticker)
  for idx in df.index:
    date = str(idx).split(' ')[0].replace('-', '')
    ticker_info[ticker][date] = Info()
    ticker_info[ticker][date].close = df.loc[idx, "종가"]
    stock.get_market_trading_volume_by_investor(date, date, ticker)    
    

# ticker_info["095570"] = {}
# df = stock.get_market_ohlcv_by_date("20210822", "20210825", "095570")
# for idx in df.index:
#   date = str(idx).split(" ")[0]
#   if date in ticker_info["095570"]:
#     ticker_info["095570"][date].close = df.loc[idx, "종가"]
#   else:
#     ticker_info["095570"][date] = Info()
#     ticker_info["095570"][date].close = df.loc[idx, "종가"]



ticker_info
```
